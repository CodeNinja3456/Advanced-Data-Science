```python
import pandas
```

```python
data=pandas.read_csv('cleanedDataset.csv',parse_dates=['bday','enrolldate','expgradate'])
data.head(10)
```

```python
#matplotlib is a package that helps in creating visual statistical graphs
#seaborn is built on top of matplotlib that provides a higher level interface for creating graphs

import matplotlib.pyplot as plt
import seaborn as sns
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
#univariante=pie chart,bargraph,histogram
#bivariante=scatterplot,boxplot,swarmplot
#multivariante=scatterplot,heatmap,...

#univariante:countplot(name of column,color)
sns.countplot(data['HowCommute'],palette='Oranges')
plt.title('Distribution of commute status')
plt.xlabel('Number of Students')
plt.ylabel('How they commute')
```

```python
#Bivariante:barplot()
import numpy
#estimator gives as a way to calculate sum.product.std
ax=sns.barplot(x='Gender',y='Math',data=data,palette='Reds',estimator=numpy.sum)
for p in ax.patches:
    txt = str(p.get_height().round(1)) + ''
    txt_x = p.get_x()
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
#multivariante ,scatterplot(),swarmplot(),violinplot()
sns.violinplot(x='Gender',y='Math',hue='State',palette='magma',data=data)
```

```python
subset=data[['Height','Weight','Math','Reading','Writing']]
sns.heatmap(subset.corr(),annot=True,cmap='Blues')
```
