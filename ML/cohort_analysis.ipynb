```python
#import libraries(pandas)
import pandas
```

```python
#do data cleaning
data=pandas.read_csv('OnlineRetail.csv',parse_dates=['InvoiceDate'],infer_datetime_format=True)
data.head(5)
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=0,subset=['Description'],inplace=True)
data.dropna(axis=0,subset=['CustomerID'],inplace=True)
```

```python
data['InvoiceNo']=data['InvoiceNo'].astype('str')
data['CustomerID']=data['CustomerID'].astype('str')
data.isnull().sum()
```

```python
#calculate the aggregation metric(sum,mean,product) indicating how many orders were placed by each customer
import numpy
n_orders=data.groupby('CustomerID')['InvoiceNo'].nunique()
n_orders
```

```python
# find the percentage(%) of customers who ordered more than once
mult_orders_perc=numpy.sum(n_orders>1)/data['CustomerID'].nunique()
mult_orders_perc
#this calculates the % of customers with multiple orders by dividing the number of customers with multiple orders by the total number of customers. nunique() calculates the total number of unique customer ID in the dataframe/dataset
print(f'{100*mult_orders_perc:.2f}% of customers ordered more than once')

# from this analysis we can see that customers have a behaviour of ordering more than once
```

```python
# we can create an sns plot to show the number of orders
import seaborn as sns
graph=sns.displot(data=n_orders)
graph.set(
    title='Distribution of Number of Orders per customer',
    xlabel='# of orders',
    ylabel='# of customers'
)
```

```python
# step1: we are going to keep only the relevant columns and drop duplicate values.
df=data[['CustomerID','InvoiceNo','InvoiceDate']].drop_duplicates()
df
```

```python
# step2:we will create a cohort column and order_month column

df['Order_Month']=df['InvoiceDate'].dt.to_period('M')
df.head(10)
```

```python
df['Cohort']=data.groupby('CustomerID')['InvoiceDate'].transform('min').dt.to_period('M')
df.head(10)
```

```python
#  then we can aggregate the data per cohort and order_month,so that we can arrange them uniquely
df_cohort=df.groupby(['Cohort','Order_Month']).agg(num_of_customers=('CustomerID','nunique')).reset_index(drop=False)
df_cohort
# .reset_index(drop=False)-it ensures both the origin grouping('Cohort','Order_Month')and the newly placed column(num_customers)are retained as columns in hte new dataframe named df_cohort
from operator import attrgetter  #this is a function that extracts specific attributes from objects
df_cohort['period_number']=(df_cohort.Order_Month -df_cohort.Cohort).apply(attrgetter('n'))
df_cohort
```

```python
# from this analysis we can see that it doesn't take long for customers to order again after their previous order
```

```python
#next step is to pivot_table the new dataframe in a way each row contains info about a given cohort and each column contains values for each period
cohort_pivot=df_cohort.pivot_table(index='Cohort',columns='period_number',values='num_of_customers')
cohort_pivot
```

```python
# to obtain the retention matrix,we need to divide the values of each row's first value 
cohort_size=cohort_pivot.iloc[:,0]
# iloc[] is used to access and manipulate data in tabular format,in this case our data is in form of pivot_table
cohort_size
retention_matrix=cohort_pivot.divide(cohort_size,axis=0)*100
retention_matrix
```

```python
# we can see that there is a sharp drop-off in the second period, on average around 80% of customers don't make purchase in the second period.A year after the first purchase their is a  drop of retention of customer purchasing
graph2=sns.heatmap(data=retention_matrix,annot=True,cmap='magma')
```
