```python
import pandas
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#label encoding:process of converting one data type to another
#float(number),string/object(text)
data=pandas.read_csv('datasets/bank_full.csv')
data
```

```python
data.drop(['duration'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['job'].fillna('unknown',inplace=True)
data['marital'].fillna('unknown',inplace=True)
data['loan'].fillna('unknown',inplace=True)

data.isnull().sum()
```

```python
#job label encoding
#step1: groupby to see the different jobs we have
data.groupby('job').size()
```

```python
#Encoding
#In python we use replace() function to encode
data['job'].replace({
    'admin.':1,
    'blue+AC0-collar':2,
    'entrepreneur':3,
    'housemaid' :4,           
    'management':5,            
    'retired' :6,              
    'self+AC0-employed' :7,    
    'services' :8,             
    'student' :9,               
    'technician' :10,          
    'unemployed' :11,           
    'unknown' :12,               
    },inplace=True)

data.groupby('job').size()
```

```python

data['marital'].replace({
    'divorced' :  1,
    'married'  :  2,
    'single'   :  3,
    'unknown'  :  4,

},inplace=True)

data.groupby('marital').size()
```

```python

data['education'].replace({
    'basic.4y' :                1,
    'basic.6y'  :               2,
    'basic.9y'   :              3,
    'high.school' :             4,
    'illiterate'  :             5,
    'professional.course' :     6,
    'university.degree' :       7,
    'unknown'  :                8,    
    },inplace=True)
data.groupby('education').size()
```

```python


data['default'].replace({
    'no' :  1,
    'unknown'  :  2,
    'yes'   :  3,
   
},inplace=True)
data.groupby('default').size()
```

```python

data['housing'].replace({
    'no' :  1,
    'unknown'  :  2,
    'yes'   :  3,
   
},inplace=True)
data.groupby('housing').size()
```

```python
data.groupby('loan').size()
data['loan'].replace({
    'no' :  1,
    'unknown'  :  2,
    'yes'   :  3,
   
},inplace=True)
data.groupby('loan').size()
```

```python


data['contact'].replace({
    'cellular' :    1,
    'telephone':    2,
   
},inplace=True)
data.groupby('contact').size()
```

```python


data['month'].replace({
    'apr' :    1,
    'aug' :    2,
    'dec'  :   3,
    'jul'  :   4,
    'jun'  :   5,
    'mar'  :   6,
    'may'   :  7,
    'nov'   :  8,
    'oct'   :  9,
    'sep'  :   10,
   
},inplace=True)
data.groupby('month').size()
```

```python


data['day_week'].replace({
    'fri'  :  1,
    'mon'  :  2,
    'thu'  :  3,
    'tue'  :  4,
    'wed'  :  5,
   
},inplace=True)
data.groupby('day_week').size()
```

```python


data['poutcome'].replace({
    'failure' :       1,
    'nonexistent':    2,
    'success' :       3,
   
},inplace=True)
data.groupby('poutcome').size()
```

```python
array=data.values

X=array[0:41190,1:13]
X
Y=array[0:41190,13]
Y

```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)
y_train
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
print('The algorithm has predicted',prediction)
print('The actual answers are',y_test)
```

```python
from sklearn.metrics import accuracy_score
check_accuracy=accuracy_score(prediction,y_test)*100
check_accuracy
```

```python
# Imports Classification Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC

# Create an Empty List and append all models, Give the Models an alias Name
# i.e we called KNeighborsClassifier() alias KNC etc
models = []
models.append(('D Trees', DecisionTreeClassifier()))
models.append(('Gaussian', GaussianNB()))
models.append(('KNC', KNeighborsClassifier()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('Gradient Boosting', GradientBoostingClassifier()))
models.append(('Linear Disc', LinearDiscriminantAnalysis()))
models.append(('Support Machines', SVC(gamma = 'auto')))

# Import Cross Validation and KFOLD
from sklearn.model_selection import cross_val_score, KFold
# Create a for loop so that each model is tested in turn
for name, model in models:
            # Here we do a 10 split K-FOLD
            kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

            # We get the results for each Fold
            cv_results = cross_val_score(model, x_train, y_train, cv = kfold,
            scoring='accuracy')

            # Get the average of all Folds
            print(name, ' Results:= ', cv_results.mean())
```
